```python
import mysql.connector
import pandas as pd
```

```python
def conectar_mysql():
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="password", 
        database="workshop_ml"
    )
    return conn
```

```python
def consultar_predicciones():
    conn = conectar_mysql()
    query = "SELECT * FROM predicciones"
    df_predicciones = pd.read_sql(query, conn)
    conn.close()
    return df_predicciones
```
